```python
i=0
while True:
    i+=1
    if i ==100:
        break
print(i)
```

```python
for i in range(10000):    # 0부터 9999까지 반복
    print(i)
    if i == 10:    # i가 100일 때
        break   
```

```python
i=1
while i<=10 :
    i+=1
    if i%2 ==1:
        continue
    print(i, end=' ')
    
```
